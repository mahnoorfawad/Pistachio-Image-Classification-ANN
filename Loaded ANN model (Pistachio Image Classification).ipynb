#Importing dependencies

```python
from tensorflow import keras
from keras.models import load_model
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

Loading pre saved model

```python
model = load_model("/content/drive/MyDrive/model")
```

```python
model.summary()
```

Loading custom images

```python
image = Image.open('/content/drive/MyDrive/image 0.jpg')
image2 = Image.open('/content/drive/MyDrive/image 1.jpg')
```

Function to show image

```python
def img_show(image):
  plt.figure(figsize = (6,4))
  plt.imshow(image)
  plt.axis("off")
  plt.show();
  print(image.format)
  print(image.size)
  print(image.mode)
```

```python
img_show(image)
```

```python
img_show(image2)
```

Function to use ANN model and perform predictions

```python
def img(image, size):
  size = size
  image = image.resize((size, size))
  x = np.array(image, dtype = np.float32 ) / 255
  x = np.expand_dims(x, axis=0)
  return x
```

```python
np_img = img(image, 180)
np_img2 = img(image2 , 180)
```

Implementing model to make predicitons

```python
pred = model.predict(np_img)
class_pred = (pred > 0.5).astype(int)
```

```python
img_show(image)
if class_pred == 0:
  print("Prediction: Kirmizi_Pistachio")
elif class_pred == 1:
  print("Prediction: Siirt_Pistachio")
```
